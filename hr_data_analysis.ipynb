```python
#removing Unnecessary Columns 
df.drop(['EmployeeCount','StandardHours','Over18','EmployeeNumber'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# Renaming columns according to snake casing for better readability and documentation

df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
#Convert the Attrition column into numeric format for easier analysis.

df['attrition_flag'] = df['attrition'].map({'Yes':1, 'No':0})
```

```python
df.duplicated().sum()
```

```python
df.dtypes
```

```python
#creating age group column

bins = [18, 30, 40, 50, 60]
labels = ['18-30','31-40','41-50','51-60']

df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
```

```python
df.head()
```

```python
#Created salary bands to convert continuous income values into meaningful business segments
df['salary_band'] = pd.cut(df['monthlyincome'],
                           bins=3,
                           labels=['Low','Medium','High'])
```

```python
df.head()
```

```python
#Created tenure groups based on years at the company
df['tenure_group'] = pd.cut(df['yearsatcompany'],
                            bins=[0,2,5,10,40],
                            labels=['0-2','3-5','6-10','10+'])
```
